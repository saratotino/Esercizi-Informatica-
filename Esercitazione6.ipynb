```python
# Esercizio 6.1.2

def count_words(string):
  #Restituisce il numero di parole prsenti nella stringa. Gestisce anche più spazi consecutivi

  # split() senza argomenti divide su qualunque quantità di spazi
  parole = string.split()
  return len(parole)

def main():
  # Chiede una stringa all'utente
  testo = input("Inserisci una frase: ")

  # Conta le parole
  numero_parole = count_words(testo)

  # Mostra il risultato
  print("Numero di parole:", numero_parole)


# Avvio del programma
main()




```

```python
# Esercizio 6.2.4

import math

#1. Diametro da AWG
def diameter(wire_gauge):

  #Restituisce il diametro in metri del filo dato il calibro AWG
  d_mm = 0.127 * 92 ** ((36 - wire_gauge) / 39) #Diametro in millimetri
  return d_mm / 1000 #Conversione in metri


#2. Resistenza filo di rame
def copper_wire_resistance(length, wire_gauge):

  #Restituisce la reistenza di un filo di rame (Ohm) dato lunghezza in metri e calibro AWG
  rho_copper = 1.678e-8 #Resistività rame (Ohm)
  d = diameter(wire_gauge)
  A = math.pi * (d ** 2) / 4 #Area sezione trasversale
  R = rho_copper * length / 4
  return R


#3: Resistenza filo alluminio
def aluminum_wire_resistance(length, wire_gauge):

  #Restituisce la resistenza di un filo in alluminio (Ohm) dato lunghezza in metri e calibro AWG
  rho_aluminum = 2.82e-8 #Resistività alluminio (Ohm)
  d = diameter(wire_gauge)
  A = math.pi * (d ** 2) / 4
  R = rho_aluminum * length / A
  return R



#4: Programma di test
if __name__ == "__main__":
  gauge = 12 #Esempio AWG
  length = 10 #10 metri

  print(f"Diametro AWG {gauge}: {diameter(gauge)*1000:.3f} mm")
  print(f"Resitenza rame ({length} m, AWG {gauge}): {copper_wire_resistance(length, gauge):.6f} Ω")
  print(f"Resitenza alluminio ({length} m, AWG {gauge}): {aluminum_wire_resistance(length, gauge):.6f} Ω")













```

```python
# Esercizio 6.2.2

# Funzione che restituisce il valore numerico di una singola lettera romana

def valore_romano(lettera):
  valori = {
      'I' : 1,
      'V' : 5,
      'X' : 10,
      'L' : 50,
      'C' : 100,
      'D' : 500,
      'M' : 1000
  }
  return valori[lettera]


# Funzione che converte un numero romano in decimale
def romano_in_decimale(s):
  totale = 0

  while len(s) > 0:

    # Caso 1: un solo carattere rimasto
    if len(s) == 1 or valore_romano(s[0]) >= valore_romano(s[1]):
      totale += valore_romano(s[0])
      s = s[1:]    # rimuove il primo carattere

    #Caso 2: notazione sottrattiva
    else:
      differenza = valore_romano(s[1]) - valore_romano(s[0])
      totale += differenza
      s = s[2:] # rimuove i primi due caratteri

  return totale


# Programma di test
if __name__ == "__main__":
  numero_romano = "MCMLXXVIII" #1978
  risultato = romano_in_decimale(numero_romano)
  print(f"{numero_romano} = {risultato}")











```
