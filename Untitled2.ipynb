```python
# esercizio 4.1.2

def visualizza_testo():
  # legge una riga di testo e visualizza: lettere maiuscole, lettere in posizioni pari,
  # stringa con vocali sostituite, conteggio delle cifre e posizioni delle vocali

  #acquisizione della stringa di input
  testo = input("Inserisci una riga di testo (può contenere lettere, numeri, simboli): ")
  print("\n" + "="*50)
  print(f"Testo acquisito: '{testo}'")
  print("\n" + "="*50)

  #variabili per i risutati
  maiuscole = ""
  posizioni_pari = ""
  testo_sostitutivo = ""
  cifre_numeriche = 0
  posizioni_vocali = []

  #Lista di riferimento per le vocali
  vocali = "AEIOUaeiou"

  #Si itera sula stringa usando l'indice (i) e il carattere (carattere)
  for i, carattere in enumerate(testo):

    #1. Identificazione delle sole lettere maiuscole
    if carattere.isupper() and carattere.isalpha():
      maiuscole += carattere

    #2. Identificazione delle lettere in posizioni pari (le posizioni pari sono 0,2,4,6,8,...)
    if i % 2 == 0:
      posizioni_pari += carattere

    #3 e 4. Sostituzione delle vocali con un underscore (_) e registrazione delle posizioni delle vocali
    if carattere in vocali:
      testo_sostitutivo += "_"
      posizioni_vocali.append(i)
    else:
      testo_sostitutivo += carattere

    #5. Conteggio delle cifre numeriche
    if carattere.isdigit():
      cifre_numeriche += 1




  # VISUALIZZAZIONE RISULTATI
  print(f"1. Lettere maiuscole: {maiuscole if maiuscole else 'Nessuna'}")
  print(f"2. Lettere in posizioni pari: {posizioni_pari if posizioni_pari else 'Nessuna'}")
  print(f"3. Testo con vocali sostituite: {testo_sostitutivo}")
  print(f"4. Conteggio cifre numeriche: {cifre_numeriche}")
  print(f"5. Posizioni delle vocali: ")

  if posizioni_vocali:
    #se la lista non è vuota, stampa i numeri
    print(f"{posizioni_vocali}")

  else:
    #se la lista è vuota, stampa 'Nessuna'
    print("Nessuna")



#Avvio del programma
visualizza_testo()

```

```python
#esercizio 4.2.2

def analizza_sequenza_numerica():
    """
    Legge una sequenza di interi (termina con input vuoto) e calcola
    somme parziali, min/max, e conteggio pari/dispari dopo ogni inserimento.
    """

    # Inizializzazione delle variabili di calcolo
    somma_parziale = 0
    valore_minimo = None  # Usiamo None per la prima acquisizione
    valore_massimo = None
    conteggio_pari = 0
    conteggio_dispari = 0

    print("Inizia l'acquisizione. Inserisci un numero intero per volta.")
    print("Premi INVIO (stringa vuota) per terminare la sequenza.")
    print("-" * 50)

    while True:
        # Richiesta dell'input come stringa
        input_utente = input("Inserisci un numero: ").strip()

        # 1. Condizione di Terminazione: stringa vuota (INVIO)
        if input_utente == "":
            print("\nSequenza terminata con l'inserimento di una stringa vuota.")
            break

        # 2. Validazione dell'Input (senza try/except):
        # Controlla se l'input è composto da cifre, considerando il segno '-'
        if input_utente.startswith('-') and input_utente[1:].isdigit():
            # Numero negativo valido
            numero = int(input_utente)
        elif input_utente.isdigit():
            # Numero positivo o zero valido
            numero = int(input_utente)
        else:
            # Input non valido (es. lettere, decimali, segni errati)
            print(f"Input non valido ('{input_utente}'). Per favore, inserisci un numero intero.")
            continue # Salta il resto del ciclo e chiede di nuovo l'input


        # --- Esecuzione dei Calcoli ---

        # I. Calcolo Somme Parziali
        somma_parziale += numero

        # II. Calcolo Minimo e Massimo
        if valore_minimo is None: # Si esegue solo alla prima acquisizione
            valore_minimo = numero
            valore_massimo = numero
        else:
            if numero < valore_minimo:
                valore_minimo = numero
            if numero > valore_massimo:
                valore_massimo = numero

        # III. Conteggio Pari e Dispari
        if numero % 2 == 0:
            conteggio_pari += 1
        else:
            conteggio_dispari += 1


        # --- Visualizzazione dei Risultati Parziali ---
        print("\n*** Risultati Parziali Aggiornati ***")

        # I. Stampa Somma Parziale
        print(f"I. Somma Parziale Totale: **{somma_parziale}**")

        # II. Stampa Minimo e Massimo
        print(f"II. Valore Minimo Acquisito: {valore_minimo}")
        print(f"    Valore Massimo Acquisito: {valore_massimo}")

        # III. Stampa Conteggi
        print(f"III. Conteggio Pari: {conteggio_pari}")
        print(f"     Conteggio Dispari: {conteggio_dispari}")
        print("-" * 50)


# Avvio del programma
analizza_sequenza_numerica()
```

```python
#esercizio 4.2.1

import random

def mossa_intelligente(n):
  # Calcola la mossa intelligente che lascia 2^k - 1 biglie
  target = [3, 7, 15, 31, 63]

  for t in target:
    if n > t:
      diff = n - t
      if 1 <= diff <= n//2:
        return diff
  return None


def turno_utente(n):
  # Turno dell'utente
  print(f"\nNel mucchio ci sono {n} biglie.")
  max_take = n // 2
  if max_take < 1:
    print("Puoi prendere solo 1 biglia. ")
    return 1

  x = input(f"Quante biglie vuoi prendere? ( 1 - {max_take}): ")

  #Controllo che l'input sia un numero
  while not x.isdigit() or not (1<= int(x) <= max_take):
    print("Input non valido, riprova.")
    x = input(f"Quante biglie vuoi prendere? ( 1 - {max_take}): ")

  return int(x)


def turno_computer(n, intelligente):
  # Turno del computer
  print(f"\nTurno del computer. Biglie rimaste: {n}")

  if intelligente:
    mossa = mossa_intelligente(n)
    if mossa is not None:
      print(f"Computer (intelligente) prende {mossa} biglie.")
      return mossa

  # Mossa stupida o forzata
  max_take = max(1, n // 2)
  mossa = random.randint(1, max_take)
  print(f"Computer (casuale) prende {mossa} biglie.")
  return mossa


def main():
  # Dimensione iniziale
  n = random.randint(10, 100)

  # Chi inizia (0 = utente, 1 = computer)
  turno = random.randint(0,1)

  # Modalità del cpmputer
  intelligente = bool(random.randint(0,1))

  print("GIOCO DELLE BIGLIE")
  print(f"Dimensione inziale del mucchio: {n}")
  print(f"{'Lutente' if turno == 0 else 'Il computer'} ha la prima mossa")
  print(f"Il computer giocherà in modalità: {'INTELLIGENTE' if intelligente else 'STUPIDA'}")

  # Ciclo di gioco
  while n > 0:
    if turno == 0:
      presa = turno_utente(n)
      n -= presa
      if n == 0:
        print("\nHai preso l'ultima biglia... HAI PERSO!")
        break

    else:
      presa = turno_computer(n, intelligente) # Corrected typo: 'prsa' to 'presa'
      n -= presa
      if n == 0:
        print("\nIl computer ha preso l'ultima biglia... HAI VINTO!")
        break

    turno = 1 - turno #cambio turno

main()
```
