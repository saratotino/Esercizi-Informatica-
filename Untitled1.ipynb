```python
def traduci_voto_numerico():
  #chiede all'utente di inserire il voto e lo converte
  voto_lettere = input('Inserisci il voto in lettere (es. A, B, ...): ').strip().upper()

  #verifica che il voto in lettere non sia vuoto
  if not voto_lettere:
    print("Errore: Nessun voto inserito")
    return

  #estrae la lettera di base (primo carattere) e l'eventuale suffisso (+ o -)
  lettera_base = voto_lettere[0]
  suffisso = voto_lettere[1:] if len(voto_lettere) > 1 else ''

  #definisce il valore numerico di base per ciascuna lettera
  valori_base = {
      'A' : 4.0,
      'B' : 3.0,
      'C' : 2.0,
      'D' : 1.0,
      'E' : 0.0
  }

  voto_numerico = None


  #1. Condizione esterna: verifica che la lettera base sia valida
  if lettera_base in valori_base:
    voto_numerico = valori_base[lettera_base] #assegna il voto base

    #2. Condizione annidata: gestisce l'eventuale suffisso (+ o -)
    if suffisso == '+':

      #+ aumenta di 0.3, tranne per A+, che resta 4.0
      if lettera_base == 'A':
        voto_numerico = 4.0

      if lettera_base != 'F':
        voto_numerico += 0.3

      else:
        print('Errore: F+ nonè valido.')
        return


    elif suffisso == '-':

        # - diminuisco di 0.3
        if lettera_base != 'F':
          voto_numerico -= 0.3

        else:
          #caso errore
          print("Errore: Il voto F- non è valido.")
          return


    elif suffisso == "":
      # Nessun suffisso, il voto numerico è quello base
      pass

    else:
      #Caso errore: suffisso non valido
      print(f"Errore: suffisso '{suffisso}' non valido")
      return


    #3. Stampa il risultato finale, se tutto è andato bene
    # Arrotonda per gestire eventuali impecisioni di floating point, ma mantiene un formato a una cifra decimale per chiarezza

    print(f"Il voto in lettere '{voto_lettere}' corrisponde al voto numerico:  ** {voto_numerico:.1f}**")


  else:
    #Caso errore: la lettera base non è valida
    print(f"Errore: La lettera base '{lettera_base}' non è un voto valido (deve essere A,B,C,D,F)")


#Chiama la funzione
traduci_voto_numerico()












```

```python
#esercizio 3.2.4
# Verifica che se un anno è bisestile (posteriore al 1582) usando un'unica espressione booleana

def is_anno_bisestile():

  #ciclo per garantire che l'input sia un numero valido e maggiore di 1582
  while True:
    input_anno = input ('Inserisci un anno (maggiore del 1582): ').strip()

    #1. Verifica che l'input sia composto solo da cifre
    if input_anno.isdigit():

      #Se è composto da cifre, converti in intero
      anno = int(input_anno)

      #2. Verifica che l'anno sia maggiore di 1582
      if anno > 1582:
        # se l'anno è valido, esci dal ciclo while
        break

      else:
        print("L'anno deve essere maggiore del 1582 per l'applicazione delle regole gregoriane.")


    else:
      print("Input non valido. Per favore, inserisci un numero intero.")


  # Un anno è bisestile se: è divisibile per 4 e non è divisibile per 100 oppure è divisibile per 400

  is_bisestile = (anno % 4 == 0 and anno % 100 != 0) or (anno % 400 == 0)

  if is_bisestile :
    print(f"L'anno {anno} è bisestile.")
  else:
    print(f"L'anno {anno} non è bisestile.")


# Chiama la funzione per eseguire il programma
is_anno_bisestile()













```

```python
#esercizio 3.2.8

def calcola_buono_spesa():
  #calcola e visualizza l'importto del buono spesa in base al denaro speso.

  #ciclo per garantire che l'inout sia un valore numerico positivo valido
  spesa = -1.0
  while spesa < 0:
    input_spesa = input("Inserisci l'importo totale speso in prodotti alimentari ($): ").strip()

    #sostituiamo la virgola con il punto
    input_spesa = input_spesa.replace(',', '.')

    #verifica se l'input è composto solo da cifre e un eventuale punto decimale
    if input_spesa.replace('.', '', 1).isdigit():

      #tenta la conversione a float(non fallirà dopo la verifica isdigit)
      spesa = float(input_spesa)

      if spesa < 0:
        print("L'importo della spesa non può essere negativo. Riprova .")

      else:
        print("Input non valido. Per favore, inserisci un numero.")
        #Forza la spesa a un valore negativo per mantenere il ciclo attivo
        spesa = -1.0

  percentuale_buono = 0.0 #percentuale di sconto

  # FASCE DI SPESA
  # meno di 10$ -> Nessun buono (0%)
  if spesa < 10.00:
    percentuale_buono = 0.0

  # da 10$ a 60$ (8%)
  elif spesa <= 60.00:
    percentuale_buono = 0.08

  # da più di 60$ a 150$ (10%)
  elif spesa <= 150.00:
    percentuale_buono = 0.10

  #da ppiù di 150$ a 210$ (12%)
  elif spesa <= 210.00:
    percentuale_buono = 0.12

  #più di 210$ (14%)
  elif spesa > 210.00:
    percentuale_buono = 0.14

  # Calcolo dell'importo del buono spesa
  importo_buono = spesa * percentuale_buono


  #Risultati
  print("\n" + "="*40)
  print(f"Spesa totale in prodotti alimentari: {spesa:.2f} $")

  if percentuale_buono > 0 :
    percentuale_stampa = int(percentuale_buono*100)
    print(f"Percentuale sconto applicata: {percentuale_stampa}%")
    print(f"Valore del buono spesa consegnato: ** ${importo_buono:.2f}%")

  else:
    print("Nessun buono spesa assegnato per questa fascia di spesa.")
    print("="*40)


# Eseguiamo la funzione
calcola_buono_spesa()











```

```python
# Esercizio 3.2.3
def determina_stagione():
  # implementa l'algoritmo dato per determinare la stagione in base al mese e al giorno

  mese = 0
  while not (1 <= mese <= 12):
    mese = int(input("Inserisci il mese (con un numero da 1 a 12): "))

  giorno = 0
  while not (1<= giorno <= 31):
    giorno = int(input("Inserisci il giorno: "))

  # primo passaggio: assegnazione per trimestre
  if 1<= mese <= 3:
    stagione = 'Winter'

  elif 4 <= mese <= 6:
    stagione = 'Spring'

  elif 7 <= mese <= 9:
    stagione = 'Summer'

  elif 10 <= mese <= 12:
    stagione = 'Fall'

  # secondo passaggio: correzione per equinozi/solsttizi
  #condizione di correzione: mese divisibile per 3 e giorno maggiore di 21

  if mese % 3 == 0 and giorno >= 21:

    if stagione == 'Winter':
      stagione = 'Spring'
    elif stagione == 'Spring':
      stagione = 'Summer'
    elif stagione == 'Summer':
      stagione = 'Fall'
    elif stagione == 'Fall':
      stagione = 'Winter'

  print(f"Per la data {giorno}/{mese}:")
  print(f"La stagione è determinata dall'algoritmo è: {stagione}")


# Eseguiamo il programma
determina_stagione()



```
