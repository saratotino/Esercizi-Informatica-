```python
# Esercizio 7.2.2

def stampa_fila(fila):
  #Stampa la fila usando _ e X
  simboli = []

  for posto in fila:
    simboli.append("X" if posto else "_")
  print(simboli)


def trova_fila_libera_più_lunga(fila):
  # Restituisce gli indici (start, end) della fila libera più lunga
  # start = primo indice libero,  end = ultimo indice libero

  max_len = 0
  best_start = None
  best_end = None

  i = 0
  n = len(fila)

  while i < n:
    if not fila[i]:
      start = i

      while i < n and not fila[i]:
        i += 1

      end = i - 1
      length = end - start + 1

      # Aggiorna la fila più lunga
      if length > max_len:
        max_len = length
        best_start = start
        best_end = end

    i += 1

  return best_start, best_end


def parcheggia(fila):
  #Trova la fila libera più lunga e parcheggia al centro
  start, end = trova_fila_libera_più_lunga(fila)
  centro = (start + end) // 2
  fila[centro] = True


def main():
  n = int(input("Numero di posti auto: "))
  fila = [False] * n # Tutti liberi

  for _ in range(n): #Al massimo si riempie tutto
    parcheggia(fila)
    stampa_fila(fila)

main()



```

```python
# Esercizio 7.2.3

import random

# Genera una configurazione casuale di pile che sommano 45
def configurazione_iniziale():
  carte = 45
  pile = []

  while carte > 0:
    x = random.randint(1, carte)
    pile.append(x)
    carte -= x

  return pile


# Stampa le pile su una riga
def stampa_pile(pile):
  for p in pile:
    print(p, end=" ")

  print()


# Applica una mossa del solitario bulgaro
def passo(pile):
  k = len(pile)
  nuove = [p - 1 for p in pile]
  nuove = [p for p in nuove if p > 0]
  nuove.append(k)
  return nuove


def configurazione_finale():
  return sorted([1, 2, 3, 4, 5, 6, 7, 8, 9])


def main():
  pile = configurazione_iniziale()
  passo_num = 0

  print("Configurazione iniziale: ")
  stampa_pile(pile)

  # Continua finchè non si raggiunge la configurazione finale
  while sorted(pile) != configurazione_finale():
    passo_num += 1
    pile = passo(pile)
    print(f"Passo {passo_num}: ")
    stampa_pile(pile)


main()









```

```python
# Esercizio 7.1.4

def leggi_sequenza():
  #Leggi numeri interi da input fino a riga vuota
  sequenza = []

  while True:
    riga = input("Inserisci un numero (o invio per terminare) : ")
    if riga == "":
      break

    #Gestione numeri relativi
    if riga.lstrip('-').isdigit():
      sequenza.append(int(riga))
    else:
      print("Inserisce solo numeri interi.")

  return sequenza


def trova_massimi_locali(seq):
  #Restituisce la lista di posizioni dei massimi locali.
  massimi = []
  n = len(seq)

  for i in range(1, n-1):
    if seq[i] > seq[i-1] and seq[i] > seq[i+1]:
      massimi.append(i) #posizione indicizzata da 0

  return massimi


def trova_due_massimi_più_vicini(massimi):
  #Restituisce le posizioni dei due massimi locali piùà vicini tra loro

  if len(massimi) < 2:
    return None

  distanza_min = float('inf')
  coppia = None

  for i in range(len(massimi)-1):
    dist = massimi[i+1] - massimi[i]

    if dist < distanza_min:
      distanza_min = dist
      coppia = (massimi[i], massimi[i+1])

  return coppia


def main():
  seq = leggi_sequenza()
  massimi = trova_massimi_locali(seq)

  if not massimi:
    print("Non ci sono massimi locali")

  else:
    # Posizioni 1-based per maggiore chiarezza
    massimi_1based = [i+1 for i in massimi]
    print("Posizioni dei massimi locali:", massimi_1based)

    if len(massimi) >= 2:
      coppia_vicina = trova_due_massimi_più_vicini(massimi)

      #Stampiamo anche 1-based
      coppia_1based = (coppia_vicina[0]+1, coppia_vicina[1]+1)
      print("Due massimi locali più vicini: ", coppia_1based)

main()












```
