```python
# Esercizio 5.1.4

biglietti = 100
acquirenti = 0

while biglietti > 0:
  print (f"Biglietti disponibili: {biglietti}")
  richiesta = int(input("Quanti biglietti vuoi acquistare? (MAX 4): "))

  # Controllo della richiesta
  if richiesta < 1 or richiesta > 4:
    print("Puoi acquistare da 1 a 4 biglietti per volta.\n")
    continue

  # Verifica diisponibilità
  if richiesta > biglietti:
    print(f"Disponibili solo {biglietti} biglietti.\n")
    continue

  # Acquisto valido
  biglietti -= richiesta
  acquirenti += 1
  print(f"Acquisto effettuato! Biglietti rimasti: {biglietti}\n")

print(f"\nBiglietti esauriti! Numero totale dii acquirenti: {acquirenti}")



```

```python
# Esercizio 5.2.2

import random

#posizione iniziale
x = 0
y = 0

print("Posizioni durante la cammianta: ")
print(x,y)

for _ in range(100):
  direzione = random.choice(["N", "S", "E", "W"])

  if direzione == "N":
    y += 1

  elif direzione == "S":
    y -= 1

  elif direzione == "E":
    x += 1

  elif direzione == "W":
    x -= 1

  print(x,y)

print("\nPosizione finale:", (x,y))




```

```python
# Esercizio 5.2.3

#Porgramma per il modello predatore-preda di Lotka-Volterra

#Parametri
A = float(input("Inserisci valore di A (Tasso crescita prede): "))
B = float(input("Inserisci valore di B (distruzione prede da predatori): "))
C = float(input("Inserisci valore di C (mortalità predatori): "))
D = float(input("Inserisci valore di D (incremento predatori da prede): "))

x = float(input("Inserisci la numerosità iniziale delle prede: "))
y = float(input("Inserisci la numerosità iniziale dei predatori: "))

n_periodi = int(input("Inserisci il numero di periodi da simulare: "))

print("\nPeriodo | Prede (x) | Predatori(y)")
print("-"*20)

for n in range(n_periodi + 1):
  print(f"{n:7d} | {x:12.4f} | {y:12.4f}")

  #calcolo delle popolazioni per il periodo successivo
  x_new = x * (1 + A - B * y)
  y_new = y * (1 - C + D * x)

  x = x_new
  y = y_new



```
