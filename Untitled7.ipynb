```python
# Esercizio 9.1.1

#1. Scambiare primo e ultimo elemento
def swap_first_last(a):
  if len(a) > 1:
    a[0], a[-1] = a[-1], a[0]


#2. Far slittare tutti gli elementi, portando l'ultimo al primo posto
def shift_right(a):
  if len(a) > 1:
    last = a.pop()
    a.insert(0, last)


#3. Sostituire con 0 tutti i valori pari
def replace_even_with_zero(a):
  for i in range(len(a)):
    if a[i] % 2 == 0:
      a[i] = 0


#4. Sostituire ogni elemento interno col maggior dei due vicini
def replace_with_max_of_neighbors(a):

  if len(a) <= 2:
    return    #Niente da modificare

  original = a[:]   #Creaimo una copia per non perdere dati
  for i in range(1, len(a) - 1):
    a[i] = max(original[i-1], original[i+1])


#5. Eliminar elemento centrale (uno o due)
def remove_middle(a):
  n = len(a)

  if n == 0:
    return

  if n % 2 == 1:
    mid = n // 2
    del a[mid]

  else:
    mid1 = n//2 -1
    del a[mid1:mid1+2]


#6. Spostare pari davanti, mantenendo ordine relativo
def move_even_first(a):
  even = [x for x in a if x % 2 == 0]
  odd = [x for x in a if x % 2 != 0]
  a[:] = even + odd


#7. Restiuire il secondo valore maggiore della lista
def second_largest(a):
  if len(a) < 2:
    return None

  #Creiamo una copia senza modificare la lista
  b = sorted(set(a), reverse = True)
  if len(b) < 2:
    return None   #Tutti uguali

  return b[1]


#8. La lista ordinata in senso crescente
def is_sorted(a):
  for i in range(len(a)-1):
    if a[i] > a[i+1]:
      return False

  return True


#9. Contiene duplicati adiacenti?
def has_adjacent_duplicates(a):
  for i in range(len(a) - 1):
    if a[i] == a[i+1]:
      return True
  return False


#10. Contiene duplicati ovunque
def has_duplicates(a):
  return len(set(a)) < len(a)


def test():
  print("=== TEST FUNZIONI ===")

  a = [1, 4, 9, 16, 25]
  print("Originale:", a)

  b = a[:]
  swap_first_last(b)
  print("1. swap_first_last: ", b)

  b = a[:]
  shift_right(b)
  print("2. shift_right:", b)

  b = a[:]
  replace_even_with_zero(b)
  print("3. replace_even_with_zero:", b)

  b = a[:]
  replace_with_max_of_neighbors(b)
  print("4. replace_with_max_of_neighbors:", b)

  b = a[:]
  remove_middle(b)
  print("5. remove_middle:", b)

  b = a[:]
  move_even_first(b)
  print("6. move_even_first:", b)

  print("7. second_largest:", second_largest(a))

  print("8. is_sorted:", is_sorted(a))

  print("9. has_adjacent_duplicates:", has_adjacent_duplicates(a))

  print("10. has_duplicates:", has_duplicates(a))


test()

```

```python
# Esercizio 9.1.4

def main():
  print("Inserisci valori reali (positii o negativi), invio per terminae: ")

  valori = []
  while True:
    s = input()
    if s == "":
      break
    valori.append(float(s))

  if len(valori) == 0:
    print("Nessun valore inserito.")
    return

  massimo = max(abs(v) for v in valori)     #Massimo in valore asoluto

  for v in valori:
    lunghezza = int(abs(v) / massimo * 40)

    if v > 0:
      #barra positiva a destra
      print(" " * 40 + "|" + "*" * lunghezza)

    elif v < 0:
      #barra negativa a sinistra
      print(" " * (40 - lunghezza) + "*" * lunghezza + "|")

    else:
      #valore zero -> solo asse
      print(" " * 40 + "|")

main()



```

```python
# Esercizio 9.1.5

def main():
  print("Inserisci le coppie (didascalia, valore).")
  print("Premi invio sulla didascalia per terminare.\n")

  etichette = []
  valori = []
  while True:
    etichetta = input("Didascalia: ")

    if etichetta.strip() == "":
      break

    valore = float(input("Valore: "))
    etichette.append(etichetta)
    valori.append(valore)


  if len(valori) == 0:
    print("Nessun dato inserito.")
    return


  massimo = max(valori)

  print("\nGrafico a barre:\n")
  for etichetta, valore in zip(etichette, valori):
    lunghezza = int((valore / massimo) * 40)
    print(f"{etichetta} {'*' * lunghezza}")


main()










```

```python
# Esercizio 9.2.2

def sillaba_iniziale(parola):
  return parola[:2] if len(parola) >= 2 else parola


def sillaba_finale(parola):
  return parola[-2:] if len(parola) >= 2 else parola


def gioca_partita():
  print("\n--- Nuova partita ---")
  print("Inserisci *, oppure una parola vuota, per terminare la partita.")

  parole_dette = []
  turno = 1 #1 o 2

  #Prima parola del giocatore 1
  parola = input(f"Giocatore {turno}, inserisci la parola inizale: ").strip()

  if parola == "*" or parola == "":
    print("Il giocatore 1 abbondona. Fine partita.\n")
    return

  parole_dette.append(parola)
  sillaba_attesa = sillaba_finale(parola)
  turno = 2 #Passa al giocatore 2

  while True:
    parola = input(f"Giocatore {turno}, parola che inizi con '{sillaba_attesa}': "). strip()

    # Abbandono o incapacità
    if parola == "*" or parola == "":
      print(f"Il giocatore {turno} si arrende.\nPartita terminata.\n")
      return

    # Parola già detta
    if parola in parole_dette:
      print(f"ERRORE: la parola '{parola}' è già stata detta!")
      print(f"Il giocatore {turno} perde.\n")
      return

    #parola valida -> aggiorna
    parole_dette.append(parola)
    sillaba_attesa = sillaba_finale(parola)

    #Cambia turno
    turno = 1 if turno == 2 else 2


def main():
  print("GIOCO DELLE PAROLE CONCATENATE")
  print("Sillabe di 2 caratteri. Per uscire dal gioco, inserisci 'stop'.\n")

  while True:
    comando = input("Vuoi iniziare una nuova partita? (s/n): ").strip().lower()

    if comando == "s":
      gioca_partita()

    elif comando == "n" or comando == "stop":
      print("Uscita dal gioco.")
      break

    else:
      print("Comando non valido.")


main()


```

```python
# Esercizio 9.2.5

def discount(prices, is_pet):
  #Conta animali e non-animali

  num_animali = 0
  prezzi_non_animali = []

  for i in range(len(prices)):
    if is_pet[i]:
      num_animali += 1
    else:
      prezzi_non_animali.append(prices[i])


  # Requisiti per ottenere lo sconto
  if num_animali >= 1 and len(prezi_non_animali) >= 5:
    return 0.20 * sum(prezzi_non_animali)
  else:
    return 0.0


def main():
  prices = []
  is_pet = []

  print("Inserisci i prodotti. Usa prezzo -1 per terminare.\n")

  while True:
    prezzo = float(input("Prezzo del prodotto: "))
    if prezzo == -1:
      break

    tipo = input("E' un animale? (Y/N): ").strip().upper()
    while tipo not in ("Y", "N"):
      tipo = input("Risposta non valida. Inserisci Y o N: ").strip().upper()

    prices.append(prezzo)
    is_pet.append(tipo == "Y")

  #Calcolo dello sconto
  s = discount(prices, is_pet)

  print("\nTotale prodotti:", len(prices))
  print("Sconto da applicare:", round(s,2), "euro")


main()



```

```python
# Esercizio 9.2.3

def name_of_best_customer(sales, customers):
  if len(sales) == 0:
    return None   #Nessun cliente

  max_index = 0
  for i in range(1, len(sales)):
    if sales[i] > sales[max_index]:
      max_index = i

  return customers[max_index]


def main():
  sales = []
  customers = []

  print("Inserisci spesa e nome del cliente. Usa 0 come sentinella per terminare.\n")

  while True:
    amount = float(input("Importo speso: "))
    if amount == 0:
      break

    name = input("Nome del cliente:")
    sales.append(amount)
    customers.append(name)
    print()

  best = name_of_best_customer(sales, customers)

  if best is None:
    print("Nessun cliente registrato oggi.")

  else:
    print("\nIl miglior cliente del giorno è:", best)

main()


```

```python
# Esercizio 9.2.6

#Una cella è inondata se la sua altezza è minore o gule al livello dell'acqua,
#altrimenti resta asciutto
def flood_map(heights, water_level):
  N = len(heights)
  flood = []

  for i in range(N):
    row = []
    for j in range(N):
      if heights[i][j] < water_level:
        row.append('*')
      else:
        row.append(' ')

    flood.append(row)

  return flood

# Lettura Matrice
N = 10
heights =[]

print("Inserisci 100 valori di altitudine: ")
for i in range(N):
  row = list(map(float, input().split()))
  heights.append(row)

min_heights = min(min(row) for row in heights)
max_heights = max(max(row) for row in heights)

step = (max_heights - min_heights) / 10


for k in range(1, 11):
  water_level = min_heights + k*step
  print(f"\nLivello dell'acqua = {water_level:.2f}")

  flood = flood_map(heights, water_level)

  for row in flood:
    print(''.join(row))






```

```python
# Esercizio 9.2.4

N = int(input("Inserisci N: "))

#Crea matrice NxN inizializzata a 0
matrix = [[0] * N for _ in range(N)]

num = 1
top, bottom = 0, N-1
left, right = 0, N-1

while num <= N*N:
  #Da sinistra a destra
  for j in range(left, right + 1):
    matrix[top][j] = num
    num += 1
  top += 1

  #Dall'alto in basso
  for i in range(top, bottom + 1):
    matrix[i][right] = num
    num += 1
  right -= 1

  #Da destra a sinistra
  for j in range(right, left - 1, -1):
    matrix[bottom][j] = num
    num += 1
  right -= 1

  #Dal basso in alto
  for i in range(bottom, top - 1, -1):
    matrix[i][left] = num
    num += 1
  left += 1

#Stampa la matrice
for row in matrix:
  for value in row:
    print(f"{value:3}", end=" ")
  print()


```
